```python
train.replace({"gender": {"male": 1, "female": 0}}, inplace=True)
train.replace({"smoker": {"yes": 1, "no": 0}}, inplace=True)
test.replace({"gender": {"male": 1, "female": 0}}, inplace=True)
test.replace({"smoker": {"yes": 1, "no": 0}}, inplace=True)
train = pd.get_dummies(train, columns=['region'])
test = pd.get_dummies(test, columns=['region'])
train.head()
```

## 6.1 - Regression without Regularization

```python
train['age'] = train['age'].apply(lambda x: x ** 2)
test['age'] = test['age'].apply(lambda x: x ** 2)
train.head()
```

```python
X_train = train.loc[:, train.columns != 'charges'].values
Y_train = train[['charges']].values
X_test = test.loc[:, test.columns != 'charges'].values
Y_test = test[['charges']].values
```

```python
w_predict = np.matmul(np.matmul(np.linalg.inv(
    np.matmul(np.transpose(X_train), X_train)), np.transpose(X_train)), Y_train)
```

```python
w_predict
```

#### 6.1 Part B

```python
def MSE(y, y_predict):
    return np.average(
        np.array(list(map(lambda x: x**2, np.subtract(y, y_predict)))))
```

```python
train_rates = []
train_errs = []
test_errs = []
for train_rate in range(20, 1000, 10):
    train_rates.append(train_rate)
    w_predict = np.matmul(np.matmul(np.linalg.inv(
        np.matmul(
            np.transpose(X_train[:train_rate]), X_train[:train_rate])),
                np.transpose(X_train[:train_rate])), Y_train[:train_rate])
    y_predict_train = np.matmul(X_train[:train_rate], w_predict)
    y_predict_test = np.matmul(X_test, w_predict)
    train_errs.append(MSE(Y_train[:train_rate], y_predict_train))
    test_errs.append(MSE(Y_test, y_predict_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(train_rates, train_errs)
plt.plot(train_rates, test_errs)
plt.ylabel('Error')
plt.xlabel('Number of train data')
plt.show()
```

## 6.2

#### Batch Gradient Descent

```python
eta = 0.1
w = np.random.rand(X_train.shape[1])
step_limit = 20000
Y_train = np.ravel(Y_train)
```

```python
test_errs = []
steps = []
step = 0
while step < step_limit:
    gradient_of_J = \
        -2 * np.matmul(np.transpose(X_train), Y_train - np.matmul(X_train, w))
    w = w - eta * (gradient_of_J / np.linalg.norm(gradient_of_J))
    y_predict_test = np.matmul(X_test, w)
    test_errs.append(MSE(Y_test, y_predict_test))
    steps.append(step)
    step += 1
w
```

```python
import matplotlib.pyplot as plt
plt.plot(steps, test_errs)
plt.ylabel('Error')
plt.xlabel('Steps')
plt.show()
```

#### Stochastic Gradient Descent

```python
eta = 0.1
w = np.random.rand(X_train.shape[1])
step_limit = 2000
```

```python
test_errs = []
steps = []
step = 0
while step < step_limit:
    for i in range(X_train.shape[0]):
        gradient_of_Ji = (Y_train[i] - np.matmul(w, X_train[i])) * X_train[i]
        w = w + eta * (gradient_of_Ji / np.linalg.norm(gradient_of_Ji))
        y_predict_test = np.matmul(X_test, w)
        test_errs.append(MSE(Y_test, y_predict_test))
        steps.append(step)
        step += 1
w
```

```python
import matplotlib.pyplot as plt
plt.plot(steps, test_errs)
plt.ylabel('Error')
plt.xlabel('Steps')
plt.show()
```

## 6.3 Regression with Regularization

```python
K = 5
pivot = int(X_train.shape[0] / K)
N = X_train.shape[0]
D = X_train.shape[1]
lambda_set = [10**-4, 10**-3, 10**-2, 10**-1, 1, 10, 10**2, 10**3, 10**4]
```

```python
def SSE(y, y_predict):
    return np.sum(np.array(list(map(lambda x: x**2, np.subtract(y, y_predict)))))
```

##### Cross Validation

```python
lambda_errs = []
for lam in lambda_set:
    error_for_lam = 0
    for i in range(5):
        folded_X_train = \
            np.concatenate((X_train[:i * pivot], X_train[(i + 1) * pivot:]))
        folded_Y_train = \
            np.concatenate((Y_train[:i * pivot], Y_train[(i + 1) * pivot:]))
        folded_X_test = X_train[i*pivot:(i+1)*pivot]
        folded_Y_test = Y_train[i*pivot:(i+1)*pivot]
        w = np.matmul(
            np.linalg.inv(np.matmul(np.transpose(folded_X_train), folded_X_train) +
                                    lam * np.identity(D)),
            np.matmul(np.transpose(folded_X_train), folded_Y_train))
        y_predict = np.matmul(folded_X_test, w)
        error_for_lam += SSE(folded_Y_test, y_predict)
    error_for_lam /= K
    lambda_errs.append(error_for_lam)    
lambda_errs
```

```python
import matplotlib.pyplot as plt
log_lambda_set = [-4, -3, -2, -1, 0, 1, 2, 3, 4]
plt.plot(log_lambda_set, test_errs)
plt.ylabel('Error')
plt.xlabel('Log Lambda')
plt.show()
```
